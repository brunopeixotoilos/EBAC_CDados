#### 1. Baixando dados de ativos
    1. Baixe os dados de três ativos da sua escolha do Yahoo Finance, correspondente a um período de pelo menos 2 anos até a data mais recente.
    2. Salve esses dados para que você possa utilizá-los na próxima tarefa.

```python
from pandas_datareader import data as web

import plotly.express as px
from plotly import graph_objects
import yfinance as yf

import pandas as pd
import numpy as np
```

```python
%%time
ativos = ['^BVSP', 'WEGE3.SA', 'GGBR4.SA', 'ALUP11.SA']
dt_ini = '2022-01-01'

# Baixar os dados com yfinance
dados = yf.download(ativos, start=dt_ini)

dados.head()
```

```python
dados.to_csv('dados.csv', index=False, encoding='utf-8')
```

#### 2. Observando a volatilidade dentro do dia
1. Calcule a amplitude de variação do valor do ativo dentro do dia. Essa amplitude você pode calcular com um comando simples de adicionar uma coluna no Pandas: (valor máximo - valor mínimo).
2. Monte um gráfico de linha mostrando a variação dessa amplitude ao longo do período coletado, por ativo.
3. Veja se consegue identificar que ativo apresentou maior volatilidade.

```python
dados.columns
```

```python
# Adicionando a coluna de amplitude

# Empilhar o MultiIndex no nível 'Ticker'
dados = dados.stack()

# Calcular a amplitude (High - Low) no formato empilhado
dados['Amplitude'] = dados['High'] - dados['Low']

# Reorganizar para o formato original com 'unstack()'
dados_final = dados.unstack()

# Exibir a estrutura resultante
dados_final.head()
```

```python
dados.head(20)
```

```python
dados_plot = dados.reset_index()

fig = px.line(dados_plot, 
              x='Date',
              y='Amplitude',
              color='Ticker',
              title='Variação da Amplitude por Ativo ao Longo do Tempo',
              template='plotly_white',
              labels={
                  'Date': 'Data',
                  'Amplitude': 'Amplitude',
                  'Ticker': 'Ativo'
              })

# Personalizando o layout
fig.update_layout(
    hovermode='x unified',
    margin=dict(l=50, r=50, t=50, b=50)
)

# Customizando o hover
fig.update_traces(
    hovertemplate="<br>".join([
        "Data: %{x}",
        "Amplitude: %{y:.2f}",
        "<extra></extra>"
    ])
)

fig.show()
```

```python
print(dados.index.names)
```

```python
# Filtrando para remover o ^BVSP
dados_filtered = dados[dados.index.get_level_values('Ticker') != '^BVSP']

# Calculando a volatilidade média
vol_media = dados_filtered.groupby('Ticker')['Amplitude'].mean().round(4)

# Criando um DataFrame para facilitar a visualização
vol_df = pd.DataFrame(vol_media).reset_index()
vol_df.columns = ['Ativo', 'Volatilidade Média']

# Exibindo a tabela
print("\nVolatilidade Média por Ativo:")
print(vol_df.to_string(index=False))

# Criando o gráfico de barras
fig = px.bar(vol_df, 
             x='Ativo', 
             y='Volatilidade Média',
             title='Volatilidade Média por Ativo',
             color='Ativo',
             template='plotly_white')

fig.update_layout(
    showlegend=False,
    yaxis_title='Volatilidade Média',
    xaxis_title='Ativo',
    # Ajustando a largura das barras
    bargap=0.5,  # Espaço entre as barras (0 a 1)
    height=500,  # Altura do gráfico
    width=800    # Largura do gráfico
)

# Customizando o hover
fig.update_traces(
    hovertemplate="<br>".join([
        "Data: %{x}",
        "Amplitude: %{y:.2f}",
        "<extra></extra>"
    ])
)

fig.show()
```

Como podemos visualizar nos gráficos, o ativo com maior volatilidade foi o da WEGE3.SA, que apresentou maior amplitude média ao longo do período analisado.
