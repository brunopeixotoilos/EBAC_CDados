#### 2. Média móvel

1. Faça uma média móvel  em 90 dias dos valores de fechamento ajustados dos ativos que você escolheu
2. Utilizando a fução ```rolling()```, calcule o desvio pardão desses ativos ao longo do tempo, utilizando a mesma janela móvel do item acima
2. Monte um gráfico de linhas para as médias móveis dos ativos da sua base
4. Monte um gráfico de linhas para os desvios padrão em janela móvel
5. Compare os dois gráficos que você fez com o da tarefa anterior. Veja se eles apontam conclusões semelhantes quanto aos períodos de maior volatilidade (variação) dos ativos.

````python
# Selecionar apenas os preços de fechamento
close_prices = dados.xs('Close', level='Price', axis=1)

# Calcular média móvel de 90 dias
ma_90 = close_prices.rolling(window=90).mean()

# Calcular desvio padrão móvel de 90 dias
std_90 = close_prices.rolling(window=90).std()

# Preparar dados para plotly express (precisa estar no formato "long"/"vertical")
ma_90_long = ma_90.reset_index().melt(id_vars='Date', var_name='Ativo', value_name='Preço')
std_90_long = std_90.reset_index().melt(id_vars='Date', var_name='Ativo', value_name='Desvio Padrão')

# Gráfico das médias móveis
fig_ma = px.line(ma_90_long, 
                 x='Date', 
                 y='Preço',
                 color='Ativo',
                 title='Média Móvel de 90 dias - Preços de Fechamento')

fig_ma.update_layout(
    template='plotly_white',
    legend=dict(
        orientation="h",
        yanchor="bottom",
        y=1.02,
        xanchor="right",
        x=1
    )
)

fig_ma.show()

# Gráfico dos desvios padrão
fig_std = px.line(std_90_long, 
                  x='Date', 
                  y='Desvio Padrão',
                  color='Ativo',
                  title='Desvio Padrão Móvel de 90 dias')

fig_std.update_layout(
    template='plotly_white',
    legend=dict(
        orientation="h",
        yanchor="bottom",
        y=1.02,
        xanchor="right",
        x=1
    )
)

fig_std.show()

# Imprimir estatísticas

print("\nEstatísticas do Desvio Padrão:")
print(std_90.describe())
````

As variações apresentadas nesses últimos gráficos foram bem menores do que no gráfico de amplitude. No entanto, as conclusões parecem ser as mesmas, onde vemos o ativo da WEGE3.SA com maior oscilação na média móvel, bem como maior desvio padrão, ou seja, o ativo apresenta maior variação e, portanto, maior volatilidade.

#### 3. Gráfico de *candlestick*
1. Selecione o período dos últimos 60 dias (corridos) para um dos ativos da sua base
2. Monte um gráfico de *candlestick* para esses dados.

````python
# Definir a data final e inicial

data_final = dados.index[-1]
data_inicial = data_final - timedelta(days=60)

# Selecionar os dados da WEGE, o com maior desvio padrão

ativo = 'WEGE3.SA'
dados_ativo = pd.DataFrame({
    'Open': dados[('Open', ativo)][data_inicial:data_final],
    'High': dados[('High', ativo)][data_inicial:data_final],
    'Low': dados[('Low', ativo)][data_inicial:data_final],
    'Close': dados[('Close', ativo)][data_inicial:data_final],
})

# Para o candlestick, precisamos usar graph_objects

fig_candle = graph_objects.Figure(
    data=[graph_objects.Candlestick(
    x=dados_ativo.index,
    open=dados_ativo['Open'],
    high=dados_ativo['High'],
    low=dados_ativo['Low'],
    close=dados_ativo['Close'],
    name=ativo
)])

# Alterações no layout

fig_candle.update_layout(
    title=f'Gráfico Candlestick - {ativo} (Últimos 60 dias)',
    yaxis_title='Preço',
    xaxis_title='Data',
    template='plotly_white'
)

fig_candle.show()
````
