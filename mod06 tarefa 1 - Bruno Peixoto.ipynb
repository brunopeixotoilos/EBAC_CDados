```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('sinasc_RO_2019.csv')

df.head()
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
#variaveis = ['APGAR1', 'APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE']

plt.figure(figsize=(10, 5))

sns.countplot(x='APGAR1', data=df)
plt.title('Contagem de valores para APGAR1')
plt.xlabel('APGAR1')
plt.ylabel('Contagem')

plt.show()
```

```python
plt.figure(figsize=(10, 5))

sns.countplot(x='APGAR5', data=df)
plt.title('Contagem de valores para APGAR5')
plt.xlabel('APGAR5')
plt.ylabel('Contagem')

plt.show()
```

```python
plt.figure(figsize=(4, 5))

sns.countplot(x='SEXO', data=df)
plt.title('Contagem de valores para SEXO')
plt.xlabel('SEXO')
plt.ylabel('Contagem')

plt.show()
```

```python
plt.figure(figsize=(10, 5))

sns.countplot(x='GESTACAO', data=df)
plt.title('Contagem de valores para GESTACAO')
plt.xlabel('GESTACAO')
plt.ylabel('Contagem')

plt.show()
```

```python
plt.figure(figsize=(10, 5))

sns.countplot(x='ESTCIVMAE', data=df)
plt.title('Contagem de valores para ESTCIVMAE')
plt.xlabel('ESTCIVMAE')
plt.ylabel('Contagem')

plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
plt.figure(figsize=(10, 5))

sns.barplot(x='CONSULTAS', y='APGAR5', data=df)
plt.title('Média de APGAR5 por CONSULTAS')
plt.xlabel('CONSULTAS')
plt.ylabel('Média de APGAR5')
plt.ylim(0, 11)

# Exiba o gráfico
plt.show()
```

```python
media_apgar5_por_consultas = df.groupby('CONSULTAS')['APGAR5'].mean()
print(media_apgar5_por_consultas)
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
fig, axes = plt.subplots(2, 2, figsize=(19, 7))

sns.countplot(data=df, x='SEXO', ax=axes[0, 0])
axes[0, 0].set_title('Frequência de SEXO')

sns.countplot(data=df, x='GESTACAO', ax=axes[0, 1])
axes[0, 1].set_title('Frequência de GESTACAO')

sns.countplot(data=df, x='ESTCIVMAE', ax=axes[1, 0])
axes[1, 0].set_title('Frequência de ESTCIVMAE')

sns.countplot(data=df, x='CONSULTAS', ax=axes[1, 1])
axes[1, 1].set_title('Frequência de CONSULTAS')

# Ajuste o layout da figura para evitar sobreposição
plt.tight_layout()

plt.show()
```
