```python
reg = smf.ols('np.log(renda) ~ sexo + posse_de_veiculo + posse_de_imovel + qtd_filhos + tipo_renda + educacao + estado_civil + tipo_residencia + idade + tempo_emprego + qt_pessoas_residencia', data = df).fit()

reg.summary()
```

```python
reg_reduzida = smf.ols('np.log(renda) ~ sexo + posse_de_veiculo + posse_de_imovel + qtd_filhos + C(tipo_renda) + estado_civil + idade + tempo_emprego + qt_pessoas_residencia', data = df).fit()

reg_reduzida.summary()
```

```python
# Criando uma coluna binária para identificar o tipo de renda = empresario
df['tipo_renda_empresario'] = df['tipo_renda'].apply(lambda x: 1 if x == 'Empresário' else 0)

# Criando uma coluna binária para identificar o tipo de renda = empresario
df['tipo_uniao'] = df['estado_civil'].apply(lambda x: 1 if x == 'União' else 0)

reg_reduzida_2 = smf.ols('np.log(renda) ~ sexo + posse_de_veiculo + posse_de_imovel + qtd_filhos + tipo_renda_empresario + tipo_uniao + idade + tempo_emprego + qt_pessoas_residencia', data = df).fit()

reg_reduzida_2.summary()
```

```python
reg_reduzida_3 = smf.ols('np.log(renda) ~ sexo + posse_de_veiculo + posse_de_imovel + qtd_filhos + tipo_renda_empresario + idade + tempo_emprego + qt_pessoas_residencia', data = df).fit()

reg_reduzida_3.summary()
```

```python
reg_reduzida_4 = smf.ols('np.log(renda) ~ sexo + posse_de_veiculo + posse_de_imovel + tipo_renda_empresario + idade + tempo_emprego + qt_pessoas_residencia', data = df).fit()

reg_reduzida_4.summary()
```
